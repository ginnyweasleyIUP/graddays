```python
import matplotlib.pyplot as plt
import pandas as pd
#import plotly.express as px
import numpy as np
import scipy.signal as signal
```

```python
data = pd.read_csv("Datasets/Monday/LR04stack.csv")
```

```python
plotdata = np.array(data)
plotx = plotdata[:,0]
ploty = plotdata[:,1]
```

```python
plt.xlabel = "Age [ka BP]"
plt.ylabel = "Foram bent d180 [per mll PDB]"
plt.plot(plotx,ploty)
```

```python
xarr = np.linspace(1,5321,5320)
xarr2 = np.linspace(4321,5321,1000)
yinterp = np.interp(xarr,plotx,ploty)
yinterp2 = np.interp(xarr2,plotx,ploty)
yinterp
plt.plot(xarr,yinterp)
plt.plot(xarr2,yinterp2)
```

```python
periox, perioy = signal.periodogram(yinterp,scaling='spectrum')
periox2, perioy2 = signal.periodogram(yinterp2,scaling='spectrum')
```

```python
plt.loglog(1/periox,perioy)
plt.loglog(1/periox2,perioy2)
```
